# Impact Analysis for Homebase Outreach for CAMBA I (Community District 303)

**Goal**: Look at historical Homebase entries to determine a baseline for Homebase entries from Community District 303 and areas serviced by CAMBA. If there is a significant lift in 303, that establishes SumAll's impact.

```python
import pandas as pd

from ggplot import *
from itertools import combinations
```

```python
#We're working with the set of Homebase records
df = pd.read_csv('hb_date_cd.csv')
df.columns.values
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
monthly_hb = df.set_index('Date').groupby('CD').resample('M', how=len).unstack(0).fillna(0)['CD'][0:15] 

#Dropped last month due to incomplete data

camba_data = monthly_hb[['CD303', 'CD308', 'CD309', 'CD317', 
                         'CD302', 'CD306', 'CD307', 'CD310', 
                         'CD311', 'CD312', 'CD313', 'CD314', 'CD315']]
camba_data
```

```python
cd_list = ['CD303', 'CD309', 'CD314', 'CD317']
```

It looks like districts 303, 309, 317, and 314 have the highest volumes of Homebase entries, so we focus on those. If you would like to change the CD's used in subsequent plots, change the entries of `cd_list`.

```python
camba_data = monthly_hb[cd_list]
camba_data["date_id"] = camba_data.index
camba_data_melt = pd.melt(camba_data, id_vars='date_id')
camba_data_melt.head()
```

```python
%matplotlib inline
```

```python
ggplot(aes(x='date_id', y='value'), data=camba_data_melt) +\
    geom_line() +\
    facet_wrap("CD", scales="fixed") +\
    scale_x_date(breaks=date_breaks('2 months'), labels='%b %Y') +\
    ggtitle("Monthly Homebase Entries (Jan 2013 - Mar 2014)")
```

We can also look at month-over-month percentage changes:

```python
camba_pct_data = {cd: camba_data[cd].pct_change()[1:] for cd in cd_list}
camba_pct = pd.DataFrame(index=camba_data.index[1:], data = camba_pct_data)
```

```python
camba_pct['date_id'] = camba_pct.index
camba_pct_melt = pd.melt(camba_pct, id_vars='date_id', var_name='CD')
camba_pct_melt.head()
```

```python
ggplot(aes(x='date_id', y='value'), data=camba_pct_melt) +\
    geom_line() +\
    facet_wrap("CD", scales="fixed") +\
    scale_x_date(breaks=date_breaks('2 months'), labels='%b %Y') +\
    scale_y_continuous(breaks='0.05', labels='percent') +\
    ggtitle("Homebase Entries by Month")
```

It seems that the number of Homebase entries per district are highly correlated. Let's see exactly how much.

```python
camba_data.corr(min_periods=4)
```

Note that CD 317 seems noticeably poorly correlated. Let's look at a moving window to see where the issue is.

```python
from itertools import combinations

correls = pd.rolling_corr_pairwise(camba_data[cd_list], 4)

correl_pairs = combinations(cd_list, r=2)
correl_df_data = {pair[0] + '_'  + pair[1] : correls.ix[:, pair[0], pair[1]].values for pair in correl_pairs}

correl_df = pd.DataFrame(index = camba_data.index, data=correl_df_data)
correl_df["date_id"] = correl_df.index

correl_df_melt = pd.melt(correl_df, id_vars = 'date_id', var_name='CD')
```

```python
ggplot(aes(x='date_id', y='value'), data=correl_df_melt)+\
    geom_line()  +\
    facet_wrap('CD', scales="fixed") +\
    scale_x_date(breaks=date_breaks('2 months'), labels='%b %Y') +\
    scale_y_continuous(breaks='0.05', labels='percent') +\
    ggtitle("Correlations between Community Districts (Rolling 4 Month Window)")
```

Note that the correlations with 317 is almost -1 for nearly every CD in later 2013 and early 2014. This exposes the root of our poor correlation with 317 in general. There is clearly some unknown behavior in the winter which is fundamentally different in CD 317. Also note that 317 operates in CAMBA I while 314 is in CAMBA II. On the other hand, 309, 314, and 317 are all directly adjacent to each other. This suggests that while we can perform reliable comparisons with 309 and 314 for impact measurements, it may be less reliable to do so with 317.

Furthermore, note how the correlations for 309 and 314 with 317 quickly approach 1 towards the spring, as expected due to their geographic proximity. In the case of extending our intervention to these areas, one of these other districts might be a good candidate for a control group.

```python
correl_pct = pd.rolling_corr_pairwise(camba_pct[cd_list], 4)

correl_pct_pairs = combinations(cd_list, r=2)
correl_pct_df_data = {pair[0] + '_'  + pair[1] : correl_pct.ix[:, pair[0], pair[1]].values for pair in correl_pct_pairs}

correl_pct_df = pd.DataFrame(index = camba_pct.index, data=correl_pct_df_data)
correl_pct_df["date_id"] = correl_pct_df.index

correl_pct_df_melt = pd.melt(correl_pct_df, id_vars = 'date_id', var_name='CD')
```

```python
ggplot(aes(x='date_id', y='value'), data=correl_pct_df_melt)+\
    geom_line()  +\
    facet_wrap('CD', scales="fixed") +\
    scale_x_date(breaks=date_breaks('2 months'), labels='%b %Y') +\
    scale_y_continuous(breaks='0.05', labels='percent') +\
    ggtitle("Correlations between % Monthly Change of Community Districts (Rolling 4 Month Window)")
```

These correlation plots show that 303 and 309 are nearly perfectly aligned in behavior even in percent month-over-month changes. On the other hand, the spring signal alignment between 309 and 314 with 317 seems to no longer exist in spite of their geographic proximity. This could well be due to different policies between different CAMBA offices.
