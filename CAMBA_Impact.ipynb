# Impact Analysis for Homebase Outreach for CAMBA I (Community District 303)

**Goal**: Look at historical Homebase entries to determine a baseline for Homebase entries from Community District 303 and areas serviced by CAMBA. If there is a significant lift in 303, that establishes SumAll's impact.

```python
import pandas as pd
from ggplot import *
from itertools import combinations
```

```python
#We're working with the set of Homebase records
df = pd.read_csv('hb_date_cd.csv')
df.columns.values
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
monthly_hb = df.set_index('Date').groupby('CD').resample('M', how=len).unstack(0).fillna(0)['CD']
camba_data = monthly_hb[['CD303', 'CD308', 'CD309', 'CD317', 
                         'CD302', 'CD306', 'CD307', 'CD310', 
                         'CD311', 'CD312', 'CD313', 'CD314', 'CD315']]
camba_data
```

It looks like districts 303, 309, 317, and 314 have the highest volumes of Homebase entries, so we focus on those.

```python
camba_data = monthly_hb[['CD303', 'CD309', 'CD314', 'CD317']]
camba_data["date_id"] = camba_data.index
camba_data_melt = pd.melt(camba_data, id_vars='date_id')
camba_data_melt.head()
```

```python
%matplotlib inline
```

```python
ggplot(aes(x='date_id', y='value'), data=camba_data_melt) +\
    geom_line() +\
    facet_wrap("CD", scales="fixed") +\
    xlab("Month") +\
    ylab("Homebase Entries") +\
    scale_x_date(breaks=date_breaks('2 months'), labels='%b %Y') +\
    ggtitle("Homebase Entries by Month")
```

We can also look at month-over-month percentage changes:

```python
camba_pct = pd.DataFrame(index=camba_data.index[1:], 
                  data = {'CD303': camba_data['CD303'].pct_change()[1:],
                          'CD309': camba_data['CD309'].pct_change()[1:],
                          'CD314': camba_data['CD314'].pct_change()[1:],
                          'CD317': camba_data['CD317'].pct_change()[1:] })
```

```python
camba_pct['date_id'] = camba_pct.index
camba_pct_melt = pd.melt(camba_pct, id_vars='date_id', var_name='CD')
camba_pct_melt.head()
```

```python
ggplot(aes(x='date_id', y='value'), data=camba_pct_melt) +\
    geom_line() +\
    facet_wrap("CD", scales="fixed") +\
    xlab("Month") +\
    ylab("Homebase Entries \% Change over Month") +\
    scale_x_date(breaks=date_breaks('2 months'), labels='%b %Y') +\
    scale_y_continuous(breaks='0.05', labels='percent') +\
    ggtitle("Homebase Entries by Month")
```

It seems that the number of Homebase entries per district are highly correlated. Let's see exactly how much.

```python
camba_data.corr(min_periods=4)
```

Note that CD 317 seems noticeably poorly correlated. Let's look at a moving window to see where the issue is.

```python
from itertools import combinations

correls = pd.rolling_corr_pairwise(camba_data[['CD303', 'CD309', 'CD314', 'CD317']], 4)

correl_pairs = combinations(['CD303', 'CD309', 'CD314', 'CD317'], r=2)
correl_df_data = {pair[0] + '_'  + pair[1] : correls.ix[:, pair[0], pair[1]].values for pair in correl_pairs}

correl_df = pd.DataFrame(index = camba_data.index, data=correl_df_data)
correl_df["date_id"] = correl_df.index

correl_df_melt = pd.melt(correl_df, id_vars = 'date_id', var_name='CD')
```

```python
ggplot(aes(x='date_id', y='value'), data=correl_df_melt)+\
    geom_line()  +\
    facet_wrap('CD', scales="fixed") +\
    xlab("Month") +\
    ylab("Correlation") +\
    scale_x_date(breaks=date_breaks('2 months'), labels='%b %Y') +\
    scale_y_continuous(breaks='0.05', labels='percent') +\
    ggtitle("Correlations between Community Districts Rolling 4 Months")
```

Note that the correlations with 317 is almost -1 for nearly every CD in later 2013 and early 2014. This exposes the root of our poor correlation with 317 in general. There is clearly some unknown behavior in the winter which is fundamentally different in CD 317. This is particularly strange since 309, 314, and 317 are all directly adjacent to each other. This suggests that while we can perform reliable comparisons with 309 and 314 for impact measurements, it may be less relable to do so with 317.

Furthermore, note how the correlations for 309 and 314 with 317 quickly approach 1 towards the spring, as expected due to their geographic proximity. In the case of extending our intervention to these areas, one of these other districts might be a good candidate for a control group.

Finally, the existence of the highly negative correlations in Winter 2013 with 317 and the uncorrelated effect of 309 and 317 in the winter while 303 and 314 became highly negative suggests a need for further research into the key features relating these areas.
